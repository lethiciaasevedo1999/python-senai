### ***Importando as bibliotecas***

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### ***Criando o dataframe***

```python
#Base de dados extraída do ambiente hadoop = Hive
df = pd.read_csv('dados.csv')
df
```

### **Explorando os dados**

```python
df.info()
```

```python
df.shape
```

```python
#Gerando um dataframe apenas da coluna dos níveis de educação
Nivelgraduacao = df['employee.education_level']
Nivelgraduacao
```

```python
#Contagem nível de graduação dos funcionários
contagemGraduacao = Nivelgraduacao.value_counts()
contagemGraduacao
```

### **Plotando gráficos**

```python
contagemGraduacao.plot(kind='bar', color='green')
plt.xlabel('Levels')
plt.ylabel('Quantia')
plt.title('Quantidade  de pessoas por nível de graduação')
```

### **Plotando gráficos com seaborn**

```python
import seaborn as sns
```

```python
sns.lineplot(x=contagemGraduacao.index, y=contagemGraduacao.values, color='green')

# Adicionando os rótulos
plt.xlabel('Levels')
plt.ylabel('Quantidade')
plt.title('Quantidade de pessoas por nível de graduação')

# Exibir o gráfico
plt.show()
```

```python
sns.barplot(x=contagemGraduacao.index, y=contagemGraduacao.values, color='green')

# Adicionando os rótulos
plt.xlabel('Levels')
plt.ylabel('Quantidade')
plt.title('Quantidade de pessoas por nível de graduação')

# Exibir o gráfico
plt.show
```

```python
sns.pointplot(x=contagemGraduacao.index, y=contagemGraduacao.values, color='green')

# Adicionando os rótulos
plt.xlabel('Levels')
plt.ylabel('Quantidade')
plt.title('Quantidade de pessoas por nível de graduação')

                               # Exibir o gráfico
plt.show
```
